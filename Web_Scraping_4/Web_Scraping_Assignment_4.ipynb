## Solution - 1

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Send a GET request to the Wikipedia page
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
response = requests.get(url)

# Create a BeautifulSoup object from the response content
soup = BeautifulSoup(response.content, "html.parser")

# Find the table containing the video details
table = soup.find("table", class_="wikitable")

ranks = []
names = []
artists = []
upload_dates = []
views = []

# Find all the rows in the table body
rows = table.tbody.find_all("tr")

# Iterate over each row and extract the video details
for row in rows:
    # Find the cells in the row
    cells = row.find_all("td")
    
    # Ensure the row has the required number of cells
    if len(cells) == 6:
        # Extract the details from the cells
        rank = cells[0].text.strip()
        name_with_wikipedia_formatting = cells[1].text.strip()
        artist = cells[2].text.strip()
        upload_date = cells[4].text.strip()
        view = cells[3].text.strip()
        
        name = name_with_wikipedia_formatting.split('"')[1] # This step is done, in order to make scraped data clean
        
        ranks.append(rank)
        names.append(name)
        artists.append(artist)
        upload_dates.append(upload_date)
        views.append(view)
        
yt_videos = {'Rank': ranks, 'Name of the song': names, 'Artist Name': artists, 'Upload Date': upload_dates, 'Views': views}
most_viewed_videos = pd.DataFrame(yt_videos)

# Print the DataFrame
most_viewed_videos.set_index('Rank', inplace=True)
most_viewed_videos
```

## Solution - 6

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Send a GET request to the URL
url = "https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare"
response = requests.get(url)
```

```python
# Create a BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table containing the details of highest selling novels
table = soup.find('table', class_='in-article sortable')
```

```python
# Initialize lists to store the scraped details
title = []
author = []
volumes_sales = []
publisher = []
genre = []

# Iterate over the rows in the table
for row in table.find_all('tr')[1:]:
    cells = row.find_all('td')
    if len(cells) >= 5:  # Check if the row has enough cells
        title.append(cells[1].text)
        author.append(cells[2].text)
        volumes_sales.append(cells[3].text)
        publisher.append(cells[4].text)
        genre.append(cells[5].text)
```

```python
# Create a DataFrame from the scraped details
data = {
    'Book Name': title,
    'Author Name': author,
    'Volumes Sold': volumes_sales,
    'Publisher': publisher,
    'Genre': genre
}
df = pd.DataFrame(data)
df
```

## Solution - 7

```python
# Send a GET request to the URL
url = "https://www.imdb.com/list/ls095964455/"
response = requests.get(url)

# Create a BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Initialize lists to store the scraped details
names = []
year_spans = []
genres = []
runtimes = []
ratings = []
votes = []

# Find the container with the TV series details
container = soup.find('div', class_='lister-list')

# Iterate over each TV series item
for item in container.find_all('div', class_='lister-item-content'):
    name = item.h3.a.text.strip()
    names.append(name)
    
    year_span = item.find('span', class_='lister-item-year').text.strip('()')
    year_spans.append(year_span)
    
    genre = item.find('span', class_='genre').text.strip()
    genres.append(genre)
    
    runtime = item.find('span', class_='runtime').text.strip()
    runtimes.append(runtime)
    
    rating = item.find('span', class_='ipl-rating-star__rating').text.strip()
    ratings.append(rating)
    
    vote = item.find('span', attrs={'name': 'nv'})['data-value']
    votes.append(int(vote))
```

```python
# Create a DataFrame from the scraped details
data = {
    'Name': names,
    'Year Span': year_spans,
    'Genre': genres,
    'Run Time': runtimes,
    'Ratings': ratings,
    'Votes': votes
}
df = pd.DataFrame(data)

# Print the DataFrame
df
```

```python
# Initialize lists to store the scraped details
dataset_names = []
data_types = []
tasks = []
attribute_types = []
no_of_instances = []
no_of_attributes = []
years = []

# Find all the dataset rows
dataset_rows = driver.find_elements(By.XPATH, "//table[@border='1']/tbody/tr")

# Iterate over each dataset row
for row in dataset_rows[1:]:
    # Extract the dataset details from each column
    columns = row.find_elements(By.TAG_NAME, "td")
    dataset_names.append(columns[0].text)
    data_types.append(columns[3].text)
    tasks.append(columns[4].text)
    attribute_types.append(columns[5].text)
    no_of_instances.append(columns[6].text)
    no_of_attributes.append(columns[7].text)
    years.append(columns[8].text)

# Create a DataFrame from the scraped details
data = {
    "Dataset Name": dataset_names,
    "Data Type": data_types,
    "Task": tasks,
    "Attribute Type": attribute_types,
    "No of Instances": no_of_instances,
    "No of Attributes": no_of_attributes,
    "Year": years,
}
df = pd.DataFrame(data)

# Print the DataFrame
df
```
